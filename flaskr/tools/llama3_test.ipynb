```python
# Example: reuse your existing OpenAI setup
from openai import OpenAI
from IPython.display import Markdown
import textwrap

# install moduls path: C:\Users\yaniv\.cache\lm-studio\models\LM Studio Community\
# start LM Studio
# select module
# start the server
# open url: http://localhost:1234/v1/models
# Point to the local server
client = OpenAI(base_url="http://localhost:1234/v1", api_key="lm-studio")


def to_markdown(text):
    text = text.replace('â€¢', '  *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

def sendServer(my_msg):
  completion = client.chat.completions.create(
    model="LM Studio Community/Meta-Llama-3-8B-Instruct-GGUF",
    messages=[
        # {"role": "system", "content": "Always answer in rhymes."},
        {"role": "user", "content": my_msg},
    ],
    temperature=0.7,
  )

  return to_markdown(completion.choices[0].message.content)
```

```python
my_msg = "What is the meaning of life in 2 points number it 1. and 2."
sendServer(my_msg)
```
